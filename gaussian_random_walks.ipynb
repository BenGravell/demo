# Gaussian random walk demo

## Package imports

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Computation

```python
# Number of timesteps
t = int(1e3) 

# Number of sample trajectories
n = 20

# Integers 1 to t
x = np.arange(t) 

# Bound of 3*standard deviations to capture ~99.7% of realizations
y = 3*np.sqrt(x)

# Sample random Gaussian walks using cumulative summation
z = np.cumsum(np.random.randn(t, n), axis=0)
```

## Plotting

```python
plt.plot(z, color='k', alpha=0.5, lw=1)
plt.fill_between(x, -y, y, color='tab:blue', alpha=0.3)
```
